```python
#Clean up data
import sqlite3
db = sqlite3.connect('database.sqlite')
cur = db.cursor()
query = 'delete from loan where funded_amnt_inv = 0;'
cur.execute(query)
db.commit()
```

```python
df = pd.read_sql_query("select loan_status, count(*) as amt from loan group by loan_status order by amt;", db)
df
```

```python
df = pd.read_sql_query("select count(*) as total_default from loan where loan_status in ('Default','Charged Off','In Grace Period') or loan_status like '%Late%';", db)
print("Total default")
df
```

```python
df = pd.read_sql_query("select count(*) as total_default from loan where loan_status like '%Fully Paid%';", db)
print("Total paid off")
df
#只有10%的坏账，不较难找
```

```python
df = pd.read_sql_query("select count(distinct emp_title) as total_job_title from loan;", db)
print("Total job categories")
df
#太多了
```

```python
df = pd.read_sql_query("select count(distinct emp_length) as job_length from loan;", db)
df
#可以用
```

```python
df = pd.read_sql_query("select home_ownership, count(home_ownership) as count from loan group by home_ownership order by count;", db)
df
#可以用 
```

```python
df = pd.read_sql_query("select count(annual_inc) as high_annul_income from loan where annual_inc > 100000;", db)
df
#可以用 
```

```python
df = pd.read_sql_query("select count(pymnt_plan) as payment_plan from loan group by pymnt_plan;", db)
df
#可以用 
```

```python
df = pd.read_sql_query("select purpose, count(purpose) as count_purpose from loan group by purpose order by count_purpose;", db)
df
#可以用 
```

```python
df = pd.read_sql_query("select pub_rec, sub_grade as count_pub_rec from loan where pub_rec>15 order by pub_rec;", db)
df
#可以用 
```

```python
df = pd.read_sql_query("select (REPLACE(term,' months','')*installment) as supposed_to_pay, total_pymnt, term, installment, (REPLACE(term,' months','')*installment)/total_pymnt as ratio, total_pymnt_inv, total_rec_int,total_rec_prncp, out_prncp, out_prncp_inv, total_rec_late_fee, funded_amnt_inv, loan_status from loan where ratio <= 1.1 and (loan_status in ('Default','In Grace Period') or loan_status like '%Late%') order by ratio;", db)
df
#如果是坏账，最多能还到本金的多少？
```

```python
df = pd.read_sql_query("select count (loan_status), (REPLACE(term,' months','')*installment)/total_pymnt as ratio from loan where ratio <= 1.2 and loan_status = 'Current' order by ratio;", db)
df
#如果是坏账，最多能还到本金的多少？
```

```python
df = pd.read_sql_query("select out_prncp from loan where funded_amnt_inv = 0;", db)
print("loan amount different from funded amount")
df
```

```python
df = pd.read_sql_query("select funded_amnt, loan_amnt from loan where funded_amnt != loan_amnt limit 50;", db)
print("loan amount different from funded amount")
df
```

```python
df = pd.read_sql_query("select count(*) from loan where funded_amnt != loan_amnt and loan_status = 'Fully Paid';", db)
print("Amount of loan eventually paid off with loan amount different from funded amount.")
df
#有可能funded和loan_amt 不一样。而且有很大的机会不一样了也会最终pay off。
```
